# Use ESM1b to create the protein embeddings.

```python
# Convert SMILES to Morgan fingerprints with nBits=2048
def smiles_to_fingerprint(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        return list(AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=2048))
    else:
        return None

train['drug_fingerprint'] = train['Drug'].apply(smiles_to_fingerprint)
val['drug_fingerprint'] = val['Drug'].apply(smiles_to_fingerprint)
test['drug_fingerprint'] = test['Drug'].apply(smiles_to_fingerprint)

# Load the tokenizer and model
hf_token = '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'  # Replace with your write permission token
tokenizer = AutoTokenizer.from_pretrained("facebook/esm1b_t33_650M_UR50S", use_auth_token=hf_token)
model = AutoModel.from_pretrained("facebook/esm1b_t33_650M_UR50S", use_auth_token=hf_token)

# Move model to GPU
model.to(device)

# Freeze the pretrained model's parameters except for ['esm.pooler.dense.bias', 'esm.pooler.dense.weight']
for name, param in model.named_parameters():
    if name not in ['pooler.dense.bias', 'pooler.dense.weight']:
        param.requires_grad = False

# Define a custom dataset
class ProteinDataset(Dataset):
    def __init__(self, sequences):
        self.sequences = sequences

    def __len__(self):
        return len(self.sequences)

    def __getitem__(self, idx):
        return self.sequences[idx]

# Convert protein sequences to embeddings using a pretrained model
def batch_sequence_to_embedding(sequences, batch_size=32, max_length=512):
    dataset = ProteinDataset(sequences)
    dataloader = DataLoader(dataset, batch_size=batch_size)

    embeddings = []
    model.eval()  # Set the model to evaluation mode

    with torch.no_grad():
        for batch in tqdm(dataloader, desc="Processing protein sequences"):
            inputs = tokenizer(batch, return_tensors="pt", padding="max_length", truncation=True, max_length=max_length).to(device)
            outputs = model(**inputs)
            batch_embeddings = outputs.last_hidden_state.mean(dim=1).cpu()
            embeddings.append(batch_embeddings)

    return torch.cat(embeddings)

# Apply the function to get protein embeddings for train, validation, and test sets
train_protein_sequences = train['Target'].tolist()
val_protein_sequences = val['Target'].tolist()
test_protein_sequences = test['Target'].tolist()

# Ensure there are no empty or invalid sequences
train_protein_sequences = [seq for seq in train_protein_sequences if seq]
val_protein_sequences = [seq for seq in val_protein_sequences if seq]
test_protein_sequences = [seq for seq in test_protein_sequences if seq]

train['protein_embedding'] = batch_sequence_to_embedding(train_protein_sequences).tolist()
val['protein_embedding'] = batch_sequence_to_embedding(val_protein_sequences).tolist()
test['protein_embedding'] = batch_sequence_to_embedding(test_protein_sequences).tolist()

```

# Build the CrossAttentionDTIPredictor model and training

```python
# Define a custom dataset for DTI
class DTI_Dataset(Dataset):
    def __init__(self, df):
        self.drug_fingerprints = df['drug_fingerprint'].values
        self.protein_embeddings = df['protein_embedding'].values
        self.labels = df['Y'].values

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        drug_fingerprint = torch.tensor(self.drug_fingerprints[idx], dtype=torch.float32)
        protein_embedding = torch.tensor(self.protein_embeddings[idx], dtype=torch.float32)
        label = torch.tensor(self.labels[idx], dtype=torch.float32)
        return drug_fingerprint, protein_embedding, label

# Create data loaders
train_dataset = DTI_Dataset(train)
val_dataset = DTI_Dataset(val)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)

# Define the Cross-Attention DTI Predictor with increased dropout
class CrossAttentionDTIPredictor(nn.Module):
    def __init__(self, pretrained_model, drug_dim=2048, protein_dim=1280, hidden_dim=512, output_dim=1, num_heads=16):
        super(CrossAttentionDTIPredictor, self).__init__()
        self.pretrained_model = pretrained_model
        self.hidden_dim = hidden_dim

        # Linear layer to project drug fingerprint to protein embedding dimension
        self.drug_projection = nn.Linear(drug_dim, protein_dim)

        # Cross-attention layer
        self.cross_attn = nn.MultiheadAttention(embed_dim=protein_dim, num_heads=num_heads)

        # Fully connected layers with layer normalization
        self.fc1 = nn.Linear(protein_dim + protein_dim, hidden_dim)
        self.ln1 = nn.LayerNorm(hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim // 2)
        self.ln2 = nn.LayerNorm(hidden_dim // 2)
        self.fc3 = nn.Linear(hidden_dim // 2, output_dim)

        self.dropout = nn.Dropout(0.2)  # Increased dropout layer

    def forward(self, drug_fingerprint, protein_embedding):
        # Ensure all tensors are on the same device
        drug_fingerprint = drug_fingerprint.to(device)
        protein_embedding = protein_embedding.to(device)

        # Project drug fingerprint to the same dimension as protein embedding
        drug_fingerprint = self.drug_projection(drug_fingerprint).unsqueeze(1)  # (batch_size, 1, protein_dim)
        protein_embedding = protein_embedding.unsqueeze(1)  # (batch_size, 1, protein_dim)

        # Apply cross-attention: query=drug, key=protein, value=protein
        attn_output, _ = self.cross_attn(drug_fingerprint, protein_embedding, protein_embedding)
        attn_output = attn_output.squeeze(1)  # Remove the sequence dimension

        # Combine drug fingerprint and protein embedding
        x = torch.cat((drug_fingerprint.squeeze(1), attn_output), dim=1)

        # Pass through fully connected layers with ReLU activations and layer normalization
        x = F.relu(self.fc1(x))
        x = self.ln1(x)  # Apply layer normalization
        x = self.dropout(x)  # Apply dropout
        x = F.relu(self.fc2(x))
        x = self.ln2(x)  # Apply layer normalization
        x = self.dropout(x)  # Apply dropout
        x = self.fc3(x)

        return x

# Initialize the enhanced model with frozen pretrained parameters
enhanced_dti_model = CrossAttentionDTIPredictor(model, num_heads=16)
enhanced_dti_model.to(device)

# Define the optimizer and scheduler for only the trainable parameters
learning_rate = 6e-6  # Lower learning rate
optimizer = AdamW(filter(lambda p: p.requires_grad, enhanced_dti_model.parameters()), lr=learning_rate, weight_decay=1e-5)
num_epochs = 200
num_training_steps = num_epochs * len(train_loader)
num_warmup_steps = int(0.1 * num_training_steps)
lr_scheduler = get_scheduler(
    "linear",
    optimizer=optimizer,
    num_warmup_steps=num_warmup_steps,
    num_training_steps=num_training_steps
)

# Store the losses for plotting
train_losses = []
val_losses = []

# Early stopping parameters
early_stopping_patience = 20
best_val_loss = float('inf')
patience_counter = 0

# Training loop
for epoch in range(num_epochs):
    enhanced_dti_model.train()
    epoch_train_loss = 0
    for batch in train_loader:
        drug_fingerprint, protein_embedding, labels = batch
        labels = labels.to(device)

        # Forward pass
        outputs = enhanced_dti_model(drug_fingerprint, protein_embedding)

        # Compute loss and normalize by batch size
        loss = torch.nn.functional.mse_loss(outputs, labels.unsqueeze(1))
        epoch_train_loss += loss.item() / len(train_loader)

        # Backward pass and optimization
        loss.backward()
        optimizer.step()
        lr_scheduler.step()
        optimizer.zero_grad()

    # Store average training loss for the epoch
    train_losses.append(epoch_train_loss)

    # Validation
    enhanced_dti_model.eval()
    epoch_val_loss = 0
    for batch in val_loader:
        drug_fingerprint, protein_embedding, labels = batch
        labels = labels.to(device)

        with torch.no_grad():
            outputs = enhanced_dti_model(drug_fingerprint, protein_embedding)
            loss = torch.nn.functional.mse_loss(outputs, labels.unsqueeze(1))
            epoch_val_loss += loss.item() / len(val_loader)

    # Store average validation loss for the epoch
    val_losses.append(epoch_val_loss)

    print(f"Epoch {epoch+1}/{num_epochs}, Training Loss: {train_losses[-1]}, Validation Loss: {val_losses[-1]}")

    # Early stopping check
    if epoch_val_loss < best_val_loss:
        best_val_loss = epoch_val_loss
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= early_stopping_patience:
            print("Early stopping triggered.")
            break


```

```python
from sklearn.metrics import mean_squared_error, r2_score, precision_recall_curve, auc
import numpy as np
from lifelines.utils import concordance_index

# Define a function to evaluate the model
def evaluate(model, loader):
    model.eval()
    true_labels = []
    predicted_labels = []

    with torch.no_grad():
        for batch in loader:
            drug_fingerprint, protein_embedding, labels = batch
            labels = labels.to(device)

            outputs = model(drug_fingerprint, protein_embedding)
            true_labels.extend(labels.cpu().numpy())
            predicted_labels.extend(outputs.squeeze().cpu().numpy())

    true_labels = scaler.inverse_transform(np.array(true_labels).reshape(-1, 1)).flatten()
    predicted_labels = scaler.inverse_transform(np.array(predicted_labels).reshape(-1, 1)).flatten()

    mse = mean_squared_error(true_labels, predicted_labels)
    rmse = np.sqrt(mse)
    ci = concordance_index(true_labels, predicted_labels)
    r2 = r2_score(true_labels, predicted_labels)
    r2m = r2_score(true_labels, predicted_labels, multioutput='uniform_average')

    threshold = 0.5  # Adjust according to your dataset
    true_labels_binary = np.where(true_labels >= threshold, 1, 0)
    predicted_labels_binary = np.where(predicted_labels >= threshold, 1, 0)

    precision, recall, _ = precision_recall_curve(true_labels_binary, predicted_labels_binary)
    aupr = auc(recall, precision)

    return mse, rmse, ci, r2, r2m, aupr

# Evaluate on validation set
val_mse, val_rmse, val_ci, val_r2, val_r2m, val_aupr = evaluate(enhanced_dti_model, val_loader)
print(f"Validation MSE: {val_mse}, RMSE: {val_rmse}, CI: {val_ci}, R²: {val_r2}, R²m: {val_r2m}, AUPR: {val_aupr}")
```
